# Script to read text file using Numpy and plot the data

```python
import numpy as np
data = np.loadtxt("effectDQ1_on_DQ2.txt", dtype = str) # reads all data as string
#it skips the lines starting from # 
```

```python
#print(data)
print(data[0][:])
```

```python
#ar = np.array(data[:,2:])  #so far data in the form of string
ar = data[:,2:].astype(np.float)  # converted into float except 1st and 2nd column
#print(ar[0:2,:])
```

```python
print(ar[1:,2])
```

```python
curDQ2 = ar[:,2]
```

```python
print(curDQ2)
```

```python
bxDQ1 = ar[:,3]
print(bxDQ1)
```

## for ploting of above array 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
plt.figure(figsize = [10,10])
#plt.subplot(1,2,1)
plt.plot(curDQ2,bxDQ1)
#plt.plot(curDQ2,byDQ1)
plt.xlabel('Current [A]',fontsize =20)
plt.ylabel('Magnetic field [G]',fontsize =20)
plt.title('Hysterisis study of magnetic quadrupole', fontsize=16)
plt.grid(True)                                 # comment this if u do not need grid
plt.savefig("hysterisis0.png")                    # png, jpg, eps, pdf as u like
plt.show()
```

```python
plt.figure(figsize = [20,5])
# DQ1
plt.subplot(1,3,1)
plt.plot(curDQ2,bxDQ1)
plt.xlabel('Current [A]',fontsize =20)
plt.ylabel('Bx1 [G]',fontsize =20)
plt.title('Hysterisis study of magnetic quadrupole', fontsize=16)
plt.grid(True)                                 # comment this if u do not need grid

plt.subplot(1,3,2)
plt.plot(curDQ2,byDQ1)
plt.xlabel('Current [A]',fontsize =20)
plt.ylabel('By1 [G]',fontsize =20)
plt.title('Hysterisis study of magnetic quadrupole', fontsize=16)
plt.grid(True)                                 # comment this if u do not need grid

plt.subplot(1,3,3)
plt.plot(curDQ2,bzDQ1)
plt.xlabel('Current [A]',fontsize =20)
plt.ylabel('Bz1 [G]',fontsize =20)
plt.title('Hysterisis study of magnetic quadrupole', fontsize=16)
plt.grid(True)                                 # comment this if u do not need grid

plt.savefig("hysterisis1.png")                    # png, jpg, eps, pdf as u like
plt.show()
```
